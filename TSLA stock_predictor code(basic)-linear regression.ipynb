```python
import pandas as pd
```

```python
df=pd.read_csv("TSLA.csv")
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df["Date"]
```

```python
df["year"]=df["Date"].dt.year
df["month"]=df["Date"].dt.month
df["day"]=df["Date"].dt.day
df
```

```python
df.drop(["Date"],axis=1)
```

```python
from sklearn import linear_model
```

```python
x=df[["year","month","day","High","Low"]]
y=df[["Open","Close"]]
```

```python
reg=linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[2020,2,3,786.140015,673.520020]])
```

```python
reg.predict([[2023,11,15,226.54,236.33]])
```
